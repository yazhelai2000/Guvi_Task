```python
#import necessary library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
import seaborn as sns
from sklearn.metrics import accuracy_score
from scipy.stats import pearsonr
import warnings
warnings.simplefilter("ignore", UserWarning)
```

```python
#importing the training data
train=pd.read_csv("trainKNN.txt",header=None)
train.columns=["id","a","b","c","d","e","f","g","h","i","glass"]
train.head(2)
```

```python
#import the test data
test=pd.read_csv("testKNN.txt",header=None)
test.columns=["id","a","b","c","d","e","f","g","h","i","glass"]
test.head(2)
```

```python
#find the pearson corrlation for all the independent feature with the dependent feature
def cor(s1,s2):
    corr,r=pearsonr(s1,s2)
    print('Pearsons correlation: %.3f' % corr)
```

```python
#separate the independent and target variable for training set with the result of pearson test
x_train=train.drop(["a","c","d","g","i","glass"],axis=1)
x_test=test.drop(["a","c","d","g","i","glass"],axis=1)
y_train=train[["glass"]]
y_test=test[["glass"]]
```

```python
#model buliding with n_neighbors=8,p=2 which is euclidean disctance
model=KNeighborsClassifier(n_neighbors=8)
model.fit(x_train,y_train)
#testing the model with the test data(new data)
model.score(x_test,y_test)
```

```python
#predicting the test sample and store it in y_pred to check for the accuracy
y_pred=model.predict(x_test)
```

```python
#accuracy for model with n_neighbors=8,p=2 which is euclidean disctance
accuracy_score(y_test,y_pred)
```

# Model buliding with the distance metrics of  minkowski with the p=2(euclidean_distance) gives 77% model score with n_neighbors=8

```python
#model buliding with n_neighbors=8,p=1 which is manhattan distance
model1=KNeighborsClassifier(n_neighbors=8,p=1)
model1.fit(x_train,y_train)
```

```python
#testing the model with the test data(new data)
model1.score(x_test,y_test)
```

```python
#predicting the test sample and store it in y_pre to check for the accuracy
y_pre=model1.predict(x_test)
```

```python
#accuracy for model with n_neighbors=8,p=2 which is manhattan disctance
accuracy_score(y_test,y_pre)
```
